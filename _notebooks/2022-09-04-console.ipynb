```java
// always start off our code with importing libraries that we will use
import java.util.Scanner; //library to take user input
import java.lang.Math; //library to use math and numbers


public class Menu { // class is defined
    // Instance Variables
    public final String DEFAULT = "\u001B[0m";  // Default Terminal Color
    public final String[][] COLORS = { // 2D Array of ANSI Terminal Colors
        {"Default",DEFAULT},
        {"Red", "\u001B[31m"}, 
        {"Green", "\u001B[32m"}, 
        {"Yellow", "\u001B[33m"}, 
        {"Blue", "\u001B[34m"}, 
        {"Purple", "\u001B[35m"}, 
        {"Cyan", "\u001B[36m"}, 
        {"White", "\u001B[37m"}, 
    };
    // 2D column location for data
    public final int NAME = 0;
    public final int ANSI = 1;  // ANSI is the "standard" for terminal codes

    // Constructor on this Object takes control of menu events and actions
    public Menu() {
        Scanner sc = new Scanner(System.in);  // using Java Scanner Object
        
        this.print();  // print Menu
        boolean quit = false;
        while (!quit) {
            try {  // scan for Input
                int choice = sc.nextInt();  // using method from Java Scanner Object
                System.out.print("" + choice + ": ");
                quit = this.action(choice);  // take action
            } catch (Exception e) {
                sc.nextLine(); // error: clear buffer
                System.out.println(e + ": Not a number, try again.");
            }
        }
        sc.close();
    }

    // Print the menu options to Terminal
    private void print() {
        //System.out.println commands below is used to present a Menu to the user. 
        System.out.println("-------------------------\n");
        System.out.println("Choose from these choices");
        System.out.println("-------------------------\n");
        System.out.println("1 - Hello World");
        System.out.println("2 - Squaring a Number");
        System.out.println("3 - ACT Score Calculator");
        System.out.println("4 - Print Colors");
        System.out.println("0 - Quit");
        System.out.println("-------------------------\n");
    }

    private void squarenumber() {
        Scanner input = new Scanner(System.in);
        double inp_number; // used int and double wrapper classes here 
        double final_number;

        System.out.print("What is your number? \n");

        inp_number = input.nextInt();

        final_number = inp_number * inp_number; // used math class

        System.out.print(inp_number + " squared is equal to " + final_number); //mutating data

    }

    private void act_calculator() {
        Scanner input = new Scanner (System.in);
        int english, math, reading, science, sum, comp_score;

        System.out.println ("Enter English Score: ");
        english = input.nextInt();
        
        System.out.println ("Enter Math Score: ");
        math = input.nextInt();

        System.out.println ("Enter Reading Score: ");
        reading = input.nextInt();

        System.out.println ("Enter Science Score: ");
        science = input.nextInt();

        input.close();

        sum = (english + math + reading + science);
        comp_score = ((english + math + reading + science) / 4);

        if (sum % 4 < 2) {
            System.out.println("Your composite score is equal to " + comp_score);
        } else if (sum % 4 == 0) {
            System.out.println("Your composite score is equal to " + comp_score);
        }
        else {
            comp_score += 1;
            System.out.println("Your composite score is equal to " + comp_score);
        }
    }


    private boolean action(int selection) {
        boolean quit = false;

        switch (selection) {
            case 0:  
                System.out.print("Thank you for using this console!\n");
                quit = true;
                break;
            case 1:
                System.out.print("Hello World!\n");
                break;
            case 2:
                squarenumber(); // instance of a class is defined
                break;    
            case 3:
                act_calculator(); // calling a method
                break;   
            case 4:
                System.out.print("Colors Used: Red, Green, Yellow, Blue, Purple, Cyan, White");
                break;
            default:
                System.out.print("Unexpected choice, try again."); // prints error message if the input is wrong
                break;
        }
        System.out.println(DEFAULT);  // resets color, gives new line
        return quit;
    }

    // Static driver/tester method
    static public void main(String[] args)  {  
        new Menu(); // starting Menu object
    }

}
Menu.main(null);
```

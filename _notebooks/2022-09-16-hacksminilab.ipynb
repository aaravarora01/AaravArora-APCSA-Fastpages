```java
class BeeLoop {
    String [][] bee;

    public BeeLoop() {
        bee = new String[][]{
                //all the bees
                {
                    "             \\     /               ",
                    "          \\    o ^ o    /          ",
                    "            \\ (     ) /            ",
                    " ____________(%%%%%%%)____________ ",
                    "(     /   /  )%%%%%%%(  \\   \\     )",
                    "(___/___/__/           \\__\\___\\___)",
                    "   (     /  /(%%%%%%%)\\  \\     )   ",
                    "    (__/___/ (%%%%%%%) \\___\\__)    ",
                    "            /(       )\\            ",
                    "          /   (%%%%%)   \\          ",
                    "               (%%%)               ",
                    "                 !                 "
                },
                {
                    "             \\     /               ",
                    "          \\    o ^ o    /          ",
                    "            \\ (     ) /            ",
                    " ____________(%%%%%%%)____________ ",
                    "(     /   /  )%%%%%%%(  \\   \\     )",
                    "(___/___/__/           \\__\\___\\___)",
                    "   (     /  /(%%%%%%%)\\  \\     )   ",
                    "    (__/___/ (%%%%%%%) \\___\\__)    ",
                    "            /(       )\\            ",
                    "          /   (%%%%%)   \\          ",
                    "               (%%%)               ",
                    "                 !                 "
                },
                {
                    "             \\     /               ",
                    "          \\    o ^ o    /          ",
                    "            \\ (     ) /            ",
                    " ____________(%%%%%%%)____________ ",
                    "(     /   /  )%%%%%%%(  \\   \\     )",
                    "(___/___/__/           \\__\\___\\___)",
                    "   (     /  /(%%%%%%%)\\  \\     )   ",
                    "    (__/___/ (%%%%%%%) \\___\\__)    ",
                    "            /(       )\\            ",
                    "          /   (%%%%%)   \\          ",
                    "               (%%%)               ",
                    "                 !                 "
                },
                {
                    "             \\     /               ",
                    "          \\    o ^ o    /          ",
                    "            \\ (     ) /            ",
                    " ____________(%%%%%%%)____________ ",
                    "(     /   /  )%%%%%%%(  \\   \\     )",
                    "(___/___/__/           \\__\\___\\___)",
                    "   (     /  /(%%%%%%%)\\  \\     )   ",
                    "    (__/___/ (%%%%%%%) \\___\\__)    ",
                    "            /(       )\\            ",
                    "          /   (%%%%%)   \\          ",
                    "               (%%%)               ",
                    "                 !                 "
                },

        };
    }


    public void BeeSong() {
        System.out.println();
        System.out.println("The Bee Buzzing Song!");

        int BeeCount = bee.length;
        for (int i = BeeCount; i >= 1; i--) 
        {

            System.out.println("The bees go buzzing " + i + " by " + i);
            System.out.println("Hoorah! Hoorah!");
            System.out.println("The little one stops to look around!");

            for (int row = 0; row < BeeCount; row++) {

                for (int col = 0; col < bee[row].length; col++) {

                    System.out.print(bee[row][col] + " ");

                    System.out.println();
                }

                System.out.println();
            }

                BeeCount -= 1;
        }

        System.out.println("And they all go flying high, in the sky,");
        System.out.println("To go back, to their hive");
        System.out.println("             THE END              ");
    }


    public static void main(String[] args)  {
        new BeeLoop().BeeSong(); 
    }

}
BeeLoop.main(null);
```

Hacks:

- This program is in more of an Imperative Programming Style. Imperative Programming is based on functions that change at each step. In this case, the function of the program is constantly changing so this is Imperative Programming.
